## 3. Sentimental Analysis

<p><b>Example: </b>Let us do a <b>sentimental analysis</b> from twitter data about how people feel about Trump in this current situation. For that <b>TextBlob</b> package to find sentiment from their post.</p>

```python
# We are using textblob to do sentimental analysis on data from twitter
# !pip install textblob
```

<p><b>Note: </b>Textblob has a sentiment property which is a namedtuple of the form Sentiment(polarity, subjectivity). The polarity score is a float within the range [-1.0, 1.0]. The subjectivity is a float within the range [0.0, 1.0] where 0.0 is very objective and 1.0 is very subjective. Based on these values we can do sentiment analysis</p>



```python
from textblob import TextBlob

trump = pd.DataFrame(columns=['User_id','Post','polarity','subjectivity'])

for status in tweepy.Cursor(api.search, q='Trump', tweet_mode="extended").items(1000):
    text = status.full_text
    analysis = TextBlob(text)
    trump = trump.append(pd.Series([status.user.screen_name, text, analysis.sentiment.polarity, 
                                    analysis.sentiment.subjectivity], index=trump.columns ), ignore_index=True)
    
    
trump 
```

```python
# We will use the polarity value to determine if the post is positive or negative, let's set thershold value to be .5

sentiment = lambda x: 'Negative' if x <0.5  else 'Positive'

trump['sentiment'] = trump['polarity'].apply(sentiment)
```

<p>We can plot a bar graph from the data obtained.</p>

```python
trump.groupby('sentiment').count().plot(kind='bar', y='User_id',figsize=(8, 6), colors=['red','green'],legend=None)

plt.title('How people feel about Trump now')
plt.xlabel('Sentiment')
plt.ylabel('No. of Users')
```

<p>To get Coordinates of the locations we got we are using an API called <b>geopy</b> to get geocodes of the location from which we can get the latitude and longitude.</p>

```python
# First let's install geopy
# !pip install geopy
```

```python
from  geopy.geocoders import Nominatim

geolocator = Nominatim(timeout=None)

list_lat = []   # create empty lists

list_long = []


for index, row in corona_tweets.iterrows():
    
    geocoded = geolocator.geocode(row['location'])
    if geocoded is not None:
        # success
        lat = geocoded.latitude
        long = geocoded.longitude
    else:
        lat = ""
        long = ""
    list_lat.append(lat)
    list_long.append(long)
    print(str(index)+": "+str(lat)+", "+str(long))
# geocode = lambda x: geolocator.geocode(x).latitude+","+geolocator.geocode(x).longitude if (geolocator.geocode(x) is not None) else None
# longitude = lambda x: geolocator.geocode(x).longitude if (geolocator.geocode(x) is not None) else ""

# corona_tweets['latitude'] = corona_tweets['location'].apply(latitude)
# corona_tweets['longitude'] = corona_tweets['location'].apply(longitude)
corona_tweets['latitude'] = list_lat
corona_tweets['longitude'] = list_long
corona_tweets.head()
```

<p>We can get the sentiment of how people feeling from their tweets using textblob</p>

```python
from textblob import TextBlob

corona = corona_tweets


sentiment = lambda x: 'Negative' if TextBlob(x).sentiment.polarity <0.5  else 'Positive'

corona['sentiment'] = corona['tweet_text'].apply(sentiment)

corona.to_csv('corona_twit.csv') # saving it to csv
corona.head() 
```

<p>We can plot the tweet based on co-ordinates on the diferent region on United States and can see whether people are positve or negative.</p>

```python
"""
Run this piece of code if the map does not render.
"""

import folium
import pandas as pd
world_map = folium.Map(location = [37.0902,-95.7129], 
                       zoom_start=4,
                       tiles = 'CartoDB dark_matter')

corona = pd.read_csv("corona_twit.csv")
corona = corona.dropna()

colors = {'Negative':'red', 'Positive':'green'}

for index, row in corona.iterrows():
    if row.latitude != '' and row.longitude != '':
        folium.CircleMarker(
                    [row.latitude,row.longitude], radius=5,
                    color = colors[row.sentiment], fill = True, fill_color='red'
                    ).add_to(world_map)


# Many people are feeling negative about corona virus as of now.
world_map
```
